```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/content/heart.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

# **Target Variable Distribution**

```python
sns.countplot(x="target",hue="target",data=df,palette='Set2', )
plt.title("Target Distribution (0 = No Disease, 1 = Disease)")
plt.xlabel("Heart Disease")
plt.ylabel("Number of People")
plt.show()
```

# **Correlation Heatmap**

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# **Age Distribution by Heart Disease**

```python
df.hist(figsize=(15, 12), bins=20, color='steelblue', edgecolor='black')
plt.suptitle("Histogram of All Features")
plt.tight_layout()
plt.show()
```

# **Age Distribution by Heart Disease**

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df,x='age',hue='target',palette='Set1',kde=True,bins=20)
plt.title("Age Distribution by Heart Disease")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# **Box Plot â€“ Cholesterol by Target**

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df,x='target',y='chol',hue="target",palette='Set3')
plt.title("Cholesterol Distribution by Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Cholesterol")
plt.show()
```

# **Pairplot of Selected Features**

```python
sns.pairplot(df[['age', 'trestbps', 'chol', 'thalach', 'target']], hue='target', palette='husl')
plt.suptitle("Pairplot of Important Features",y=1.02)
plt.show()
```

## **NumPy Stats Example (Cholesterol)**

```python
chol_vals = df['chol'].values
print("Mean:", np.mean(chol_vals))
print("Median:", np.median(chol_vals))
print("Std Dev:", np.std(chol_vals))
print("25th Percentile:", np.percentile(chol_vals, 25))
print("75th Percentile:", np.percentile(chol_vals, 75))
```
